```python
import pandas as pd

crispr_df = pd.read_csv("depmap/crispr/CRISPRGeneEffect.csv", index_col=0)
crispr_df
```

```python
crispr_df
```

```python
model_df = pd.read_csv("depmap/Model.csv")
tissue_effects = pd.DataFrame()
for tissue, group in model_df.groupby("SampleCollectionSite"):
    try:
        selected = [x for x in group["ModelID"].values if x in crispr_df.index.values]
        cdf = crispr_df.loc[selected]
        gene_effects = cdf.mean(axis=0) < -1
        tissue_effects = tissue_effects.append(gene_effects.to_frame().reset_index().assign(tissue=tissue))
    except KeyError:
        print(tissue)
        pass
tissue_effects = (tissue_effects
                  .rename(columns={"index":"gene", 0: "is_essential"})
                  .assign(gene=lambda df: df["gene"].apply(lambda x: x.split()[0]),
                          ensg=lambda df: df["gene"].apply(lambda x: GENE2ENSEMBLE[x.split()[0]] if x.split()[0] in GENE2ENSEMBLE else None)
                         )
                 )
tissue_effects
```

```python
tissue_effects[tissue_effects["is_essential"]]["tissue"].unique()
```

```python
tissue_effects.to_csv("depmap/crispr/tissue_effect_labels.csv")
```
